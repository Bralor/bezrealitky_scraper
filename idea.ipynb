```python
def write_data_into_json(data, cache = None):
    if not cache:
        cache = list()

    for item in data:
        key, vals = collect_data(item)
        cache.append({key: vals})
    
    return cache
```

```python
def collect_data(output: dict,
                 basename: str = "https://www.bezrealitky.cz/nemovitosti-byty-domy/"
                ) -> dict:
    return output["advertEstateOffer"][0]["id"], {
        "name": f"{output['uri'].split('-', maxsplit=1)[1]}",
        "price": output["advertEstateOffer"][0]["price"],
        "currency": output["advertEstateOffer"][0]["currency"],
        "surface": output["advertEstateOffer"][0]["surface"],
        "url": f"{basename}{output['uri']}",
        "gps": output["advertEstateOffer"][0]["gps"],
        "dispozition": output["advertEstateOffer"][0]["keyDisposition"]
    }
```

```python
def create_json(data, filename):
    if os.path.exists(filename):
        with open(filename, "r+") as json_f:
            content = json.load(json_f)
        print(f"Updating existing json file {filename}")
        
        for item in data:
            if item not in content:
                json.dump(data, json_f, indent=4)
        print(f"Json {filename} updated")
    
    else:
        with open(filename, "w") as json_f:
            json.dump(data, json_f, indent=4)
        print(f"New json file created {filename}")
```
